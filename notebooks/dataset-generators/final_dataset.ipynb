```python
import os
from pathlib import Path
import pandas as pd
```

```python
parent_dir = Path.cwd().parent.parent
```

```python
# Import datasets (google trends will not be used for the regression analysis)

# Monthly
fin_gig_m_df = pd.read_csv(os.path.join(parent_dir, 'data', 'yfinance', 'gig_yfinance_monthly.csv')) 
fin_sp500_m_df = pd.read_csv(os.path.join(parent_dir, 'data', 'yfinance', 'sp500_yfinance_monthly.csv'))
fred_m_df = pd.read_csv(os.path.join(parent_dir, 'data', 'fred', 'fred_monthly_data.csv'))
google_m_df = pd.read_csv(os.path.join(parent_dir, 'data', 'google_trends', 'google_trends_monthly.csv'))

# Quarter
fin_gig_q_df = pd.read_csv(os.path.join(parent_dir, 'data', 'yfinance', 'gig_yfinance_quarterly.csv'))
fin_sp500_q_df = pd.read_csv(os.path.join(parent_dir, 'data', 'yfinance', 'sp500_yfinance_quarterly.csv'))
fred_q_df = pd.read_csv(os.path.join(parent_dir, 'data', 'fred', 'fred_quarterly_data.csv'))
google_q_df = pd.read_csv(os.path.join(parent_dir, 'data', 'google_trends', 'google_trends_quarterly.csv'))

# Annual
fin_gig_a_df = pd.read_csv(os.path.join(parent_dir, 'data', 'yfinance', 'gig_yfinance_annually.csv'))
fin_sp500_a_df = pd.read_csv(os.path.join(parent_dir, 'data', 'yfinance', 'sp500_yfinance_annually.csv'))
fred_a_df = pd.read_csv(os.path.join(parent_dir, 'data', 'fred', 'fred_annual_data.csv'))
google_a_df = pd.read_csv(os.path.join(parent_dir, 'data', 'google_trends', 'google_trends_annual.csv'))
```

```python
# Merge datasets
# Merging with the fred datasets being the driver because they go back the farthest
# Note: For simplicity, we are going to drop rows with missing data

# Monthly
monthly_gig_df = fred_m_df.merge(fin_gig_m_df, left_on=['month','year'], right_on=['Month','Year'], how='left', suffixes=('', '_drop')).dropna().merge(google_m_df, on=['month','year'], how='left', suffixes=('', '_drop')).dropna()
monthly_sp500_df = fred_m_df.merge(fin_sp500_m_df, left_on=['month','year'], right_on=['Month','Year'], how='left', suffixes=('', '_drop')).dropna().merge(google_m_df, on=['month','year'], how='left', suffixes=('', '_drop')).dropna()

# Quarterly
quarterly_gig_df = fred_q_df.merge(fin_gig_q_df, left_on=['quarter','year'], right_on=['Quarter','Year'], how='left', suffixes=('', '_drop')).dropna().merge(google_q_df, on=['quarter','year'], how='left', suffixes=('', '_drop')).dropna()
quarterly_sp500_df = fred_q_df.merge(fin_sp500_q_df, left_on=['quarter','year'], right_on=['Quarter','Year'], how='left', suffixes=('', '_drop')).dropna().merge(google_q_df, on=['quarter','year'], how='left', suffixes=('', '_drop')).dropna()

# Annual
annual_gig_df = fred_a_df.merge(fin_gig_a_df, left_on=['year'], right_on=['Year'], how='left', suffixes=('', '_drop')).dropna().merge(google_a_df, on=['year'], how='left', suffixes=('', '_drop')).dropna()
annual_sp500_df = fred_a_df.merge(fin_sp500_a_df, left_on=['year'], right_on=['Year'], how='left', suffixes=('', '_drop')).dropna().merge(google_a_df, on=['year'], how='left', suffixes=('', '_drop')).dropna()
```

```python
# Drop _drop columns

# Monthly
monthly_gig_df = monthly_gig_df[[col for col in monthly_gig_df.columns if not col.endswith('_drop')]]
monthly_sp500_df = monthly_sp500_df[[col for col in monthly_sp500_df.columns if not col.endswith('_drop')]]

# Quarterly
quarterly_gig_df = quarterly_gig_df[[col for col in quarterly_gig_df.columns if not col.endswith('_drop')]]
quarterly_sp500_df = quarterly_sp500_df[[col for col in quarterly_sp500_df.columns if not col.endswith('_drop')]]

# Annual
annual_gig_df = annual_gig_df[[col for col in annual_gig_df.columns if not col.endswith('_drop')]]
annual_sp500_df = annual_sp500_df[[col for col in annual_sp500_df.columns if not col.endswith('_drop')]]    
```

```python
# Add target variable which will be Close / CPI

# Monthly
monthly_gig_df['target'] = monthly_gig_df['Close'] / monthly_gig_df['cpi_data_monthly']
monthly_sp500_df['target'] = monthly_sp500_df['Close'] / monthly_sp500_df['cpi_data_monthly']

# Quarterly
quarterly_gig_df['target'] = quarterly_gig_df['Close'] / quarterly_gig_df['cpi_data_monthly']
quarterly_sp500_df['target'] = quarterly_sp500_df['Close'] / quarterly_sp500_df['cpi_data_monthly']

# Annual
annual_gig_df['target'] = annual_gig_df['Close'] / annual_gig_df['cpi_data_monthly']
annual_sp500_df['target'] = annual_sp500_df['Close'] / annual_sp500_df['cpi_data_monthly']
```

```python
# Write all datasets to file

# Monthly
monthly_gig_df.to_csv(os.path.abspath(os.path.join(parent_dir, './data/final/gig_monthly_data.csv')), index=False)
monthly_sp500_df.to_csv(os.path.abspath(os.path.join(parent_dir, './data/final/sp500_monthly_data.csv')), index=False)

# Quarterly
quarterly_gig_df.to_csv(os.path.abspath(os.path.join(parent_dir, './data/final/gig_quarterly_data.csv')), index=False)
quarterly_sp500_df.to_csv(os.path.abspath(os.path.join(parent_dir, './data/final/sp500_quarterly_data.csv')), index=False)

# Annual
annual_gig_df.to_csv(os.path.abspath(os.path.join(parent_dir, './data/final/gig_annual_data.csv')), index=False)
annual_sp500_df.to_csv(os.path.abspath(os.path.join(parent_dir, './data/final/sp500_annual_data.csv')), index=False)
```
