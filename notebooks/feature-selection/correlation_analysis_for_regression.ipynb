```python
import os
from pathlib import Path
import pandas as pd
import statsmodels.api as sm
```

```python
parent_dir = Path.cwd().parent.parent
```

```python
# Import datasets

# Monthly
gig_m_df = pd.read_csv(os.path.join(parent_dir, 'data', 'final', 'gig_monthly_data.csv'))
sp500_m_df = pd.read_csv(os.path.join(parent_dir, 'data', 'final', 'sp500_monthly_data.csv')) 
```

```python
columns_to_drop = ['date', 'year', 'month', 'quarter', 'Date', 'Month', 'Quarter', 'Year', 'Company', 'Ticker', 'industryKey', 'sectorKey', 'Close']
columns_to_keep = ['cpi_data_monthly',
'rec_smooth_prob_data_monthly',
'rec_nber_data_monthly',
'unemployment_data_monthly',
'labor_force_participation_data_monthly',
'job_hires_data_monthly',
'gdp_real_data_quarterly',
'm2_real_data_monthly',
'interest_rate_mtg_data_weekly',
'median_income_data_annual', # needs to be divided by cpi if it will be used in the regression
'consumer_sentiment_data_monthly',
'target'
]
```

```python
# Cleaning up the variables if you decide to hit Run All without restarting the kernel
gbl_lst = list(globals().keys())
if 'gig_m_cor_df' in gbl_lst:
    del gig_m_cor_df
if 'sp500_m_cor_df' in gbl_lst:
    del sp500_m_cor_df
```

```python
# Run correlation analysis

# Monthly
for col in sp500_m_df.columns:
    gbl_lst = list(globals().keys())
    if col in columns_to_keep:
        acf = sm.tsa.acf(sp500_m_df[col].dropna(), nlags=12)
        pacf = sm.tsa.pacf(sp500_m_df[col].dropna(), nlags=12)
        for col2 in sp500_m_df.columns:
            if col2 in columns_to_keep and col2 != col:
                ccf = sm.tsa.ccf(sp500_m_df[col].dropna(), sp500_m_df[col2].dropna(), nlags=12)
                if 'sp500_m_cor_df' not in gbl_lst:
                    sp500_m_cor_df = pd.DataFrame({'type': 'sp500', 'column': col, 'ccf_comp_col': col2
                                        , 'ccf_lag_0': ccf[0], 'ccf_lag_1': ccf[1], 'ccf_lag_2': ccf[2], 'ccf_lag_3': ccf[3], 'ccf_lag_4': ccf[4], 'ccf_lag_5': ccf[5], 'ccf_lag_6': ccf[6], 'ccf_lag_7': ccf[7], 'ccf_lag_8': ccf[8], 'ccf_lag_9': ccf[9], 'ccf_lag_10': ccf[10], 'ccf_lag_11': ccf[11]
                                        , 'acf_lag_1': acf[1], 'acf_lag_2': acf[2], 'acf_lag_3': acf[3], 'acf_lag_4': acf[4], 'acf_lag_5': acf[5], 'acf_lag_6': acf[6], 'acf_lag_7': acf[7], 'acf_lag_8': acf[8], 'acf_lag_9': acf[9], 'acf_lag_10': acf[10], 'acf_lag_11': acf[11], 'acf_lag_12': acf[12]
                                        , 'pacf_lag_1': pacf[1], 'pacf_lag_2': pacf[2], 'pacf_lag_3': pacf[3], 'pacf_lag_4': pacf[4], 'pacf_lag_5': pacf[5], 'pacf_lag_6': pacf[6], 'pacf_lag_7': pacf[7], 'pacf_lag_8': pacf[8], 'pacf_lag_9': pacf[9], 'pacf_lag_10': pacf[10], 'pacf_lag_11': pacf[11], 'pacf_lag_12': pacf[12]
                                        }, index=[0])
                else:
                    df2 = pd.DataFrame({'type': 'sp500', 'column': col, 'ccf_comp_col': col2
                                        , 'ccf_lag_0': ccf[0], 'ccf_lag_1': ccf[1], 'ccf_lag_2': ccf[2], 'ccf_lag_3': ccf[3], 'ccf_lag_4': ccf[4], 'ccf_lag_5': ccf[5], 'ccf_lag_6': ccf[6], 'ccf_lag_7': ccf[7], 'ccf_lag_8': ccf[8], 'ccf_lag_9': ccf[9], 'ccf_lag_10': ccf[10], 'ccf_lag_11': ccf[11]
                                        , 'acf_lag_1': acf[1], 'acf_lag_2': acf[2], 'acf_lag_3': acf[3], 'acf_lag_4': acf[4], 'acf_lag_5': acf[5], 'acf_lag_6': acf[6], 'acf_lag_7': acf[7], 'acf_lag_8': acf[8], 'acf_lag_9': acf[9], 'acf_lag_10': acf[10], 'acf_lag_11': acf[11], 'acf_lag_12': acf[12]
                                        , 'pacf_lag_1': pacf[1], 'pacf_lag_2': pacf[2], 'pacf_lag_3': pacf[3], 'pacf_lag_4': pacf[4], 'pacf_lag_5': pacf[5], 'pacf_lag_6': pacf[6], 'pacf_lag_7': pacf[7], 'pacf_lag_8': pacf[8], 'pacf_lag_9': pacf[9], 'pacf_lag_10': pacf[10], 'pacf_lag_11': pacf[11], 'pacf_lag_12': pacf[12]
                                        }, index=[0])
                    sp500_m_cor_df = pd.concat([sp500_m_cor_df, df2], ignore_index=True)
sp500_m_cor_df = sp500_m_cor_df.drop_duplicates().reset_index(drop=True)
```

```python
# I manually analyzed the data using the S&P 500 since it is a more diverse set and determined in the cell below which columns to keep
sp500_m_cor_df[(sp500_m_cor_df['ccf_lag_0'].abs() >= 0.7 ) | (sp500_m_cor_df['ccf_lag_1'].abs() >= 0.7 ) | (sp500_m_cor_df['ccf_lag_2'].abs() >= 0.7 ) | (sp500_m_cor_df['ccf_lag_3'].abs() >= 0.7 ) | (sp500_m_cor_df['ccf_lag_4'].abs() >= 0.7 ) | (sp500_m_cor_df['ccf_lag_5'].abs() >= 0.7 ) | (sp500_m_cor_df['ccf_lag_6'].abs() >= 0.7 ) | (sp500_m_cor_df['ccf_lag_7'].abs() >= 0.7 ) | (sp500_m_cor_df['ccf_lag_8'].abs() >= 0.7 ) | (sp500_m_cor_df['ccf_lag_9'].abs() >= 0.7 ) | (sp500_m_cor_df['ccf_lag_10'].abs() >= 0.7 ) | (sp500_m_cor_df['ccf_lag_11'].abs() >= 0.7 )
               ].sort_values(by=['column','ccf_comp_col'
                                 ], ascending=False
                            ).reset_index(drop=True)
```
